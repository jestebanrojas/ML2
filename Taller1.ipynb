```python
#Script to create a random values Rectangular Matrix.

# __init__.py

import numpy as np

array = np.random.randint(50, size=(10, 10))

print("Next is displayed the random rectangular Matrix:\n")
print(array)

print("\n\n The rank of the Matrix array is:\n")
print(np.linalg.matrix_rank(array))


try:
    detArray=np.linalg.det(array)
    print(detArray)
except Exception as err:
    print("\n\n Oops!  It was not able to calculate the matrix determinant, it is not a square matrix")


# Calculating the inverse of the matrix
invArray=np.linalg.inv(array)
print(invArray)
```
